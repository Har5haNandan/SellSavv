### Live Price Hygeine

```python
df_input['Live Price_400013'] = df_input['Live Price_400013'].str.strip("â‚¹").str.replace(",","")
df_input['Live Price_600005'] = df_input['Live Price_600005'].str.strip("â‚¹").str.replace(",","")
df_input['Live Price_122102'] = df_input['Live Price_122102'].str.strip("â‚¹").str.replace(",","")
df_input['Live Price_700016'] = df_input['Live Price_700016'].str.strip("â‚¹").str.replace(",","")
df_input['Live Price_560068'] = df_input['Live Price_560068'].str.strip("â‚¹").str.replace(",","")

df_input['Price Rule'] = df_input['Price Rule'].astype('float')

df_input['Live Price_400013'] = df_input['Live Price_400013'].astype('float')
df_input['Live Price_600005'] = df_input['Live Price_600005'].astype('float')
df_input['Live Price_122102'] = df_input['Live Price_122102'].astype('float')
df_input['Live Price_700016'] = df_input['Live Price_700016'].astype('float')
df_input['Live Price_560068'] = df_input['Live Price_560068'].astype('float')

df_input['Live Price_400013_Score'] = (df_input['Live Price_400013'] == df_input['Price Rule']).astype("int", errors='ignore')
df_input['Live Price_600005_Score'] = (df_input['Live Price_600005'] == df_input['Price Rule']).astype("int", errors='ignore')
df_input['Live Price_122102_Score'] = (df_input['Live Price_122102'] == df_input['Price Rule']).astype("int", errors='ignore')
df_input['Live Price_700016_Score'] = (df_input['Live Price_700016'] == df_input['Price Rule']).astype("int", errors='ignore')
df_input['Live Price_560068_Score'] = (df_input['Live Price_560068'] == df_input['Price Rule']).astype("int", errors='ignore')

df_input['Price'] = (df_input['Live Price_400013_Score']+df_input['Live Price_600005_Score']+
                     df_input['Live Price_122102_Score']+df_input['Live Price_700016_Score']+df_input['Live Price_560068_Score'])*20
```

### Sold By Hygeine

```python
df_input['Sold By_400013_Score'] = (df_input['Sold By_400013'] == df_input['Buy Box Rule']).astype("int")
df_input['Sold By_600005_Score'] = (df_input['Sold By_600005'] == df_input['Buy Box Rule']).astype("int")
df_input['Sold By_122102_Score'] = (df_input['Sold By_122102'] == df_input['Buy Box Rule']).astype("int")
df_input['Sold By_700016_Score'] = (df_input['Sold By_700016'] == df_input['Buy Box Rule']).astype("int")
df_input['Sold By_560068_Score'] = (df_input['Sold By_560068'] == df_input['Buy Box Rule']).astype("int")

df_input['Sold By'] = (df_input['Sold By_400013_Score']+df_input['Sold By_600005_Score']+
                       df_input['Sold By_122102_Score']+df_input['Sold By_700016_Score']+df_input['Sold By_560068_Score'])*20
```

```python
df_input['Coupon Validation'] = df_input['Live Coupon'] == df_input['Coupon Rule']
df_input['SNS Validation'] = df_input['Live SNS'] == df_input['SNS Rule']
df_input['BXGY Validation'] = df_input['Live BXGY'] == df_input['BXGY Rule']

df_input["Catalog Score"] = (df_input['Ratings'] >= 4).astype(int) + (df_input['Title Length'] >= 180).astype(int) + (df_input['Bullet Point Count'] > 5).astype(int) + (df_input['Images Count'] >= 7).astype(int) + (df_input['A+'] == 'Yes').astype(int)
df_input["Catalog Score"] = df_input["Catalog Score"]*20
df_input["Date"] = datetime.today().strftime('%Y-%m-%d')
df_input['Total Ratings'] = (
    df_input['Total Ratings']
    .str.replace(',', '', regex=True)  # Remove commas
    .str.extract(r'(\d+)')  # Extract numeric values
    .astype(float)  # Convert to float
)
```

```python
def calculate_hygiene_metrics(df):
    """
    Calculate hygiene metrics for each row in the dataset
    
    Parameters:
    df (pandas.DataFrame): DataFrame containing the required columns
    
    Returns:
    pandas.DataFrame: Original DataFrame with additional hygiene metrics columns
    """
    # Create a copy of the DataFrame to avoid modifying the original
    result_df = df.copy()
    
    # 1. Activation Hygiene (100% if both validations are True, 0% if either is False)
    result_df['Activation_Hygiene'] = ((result_df['SNS Validation'] == True) & 
                                     (result_df['BXGY Validation'] == True)) * 100
    
    
    # 2. EDD Hygiene (100% if EDD â‰¤ 2, 0% if EDD > 2), Price Hygiene, Buy Box_Hygiene
    result_df['EDD_Hygiene'] = result_df['EDD']
    result_df['Price_Hygiene'] = result_df['Price']
    result_df['Buy Box_Hygiene'] = result_df['Sold By']
    
    
    # 3. Catalog Hygiene (use raw Catalog Score for each row)
    result_df['Catalog_Hygiene'] = result_df['Catalog Score']
    
    # 4. Rating Hygiene (divide by 5 and multiply by 100 for percentage)
    result_df['Rating_Hygiene'] = (result_df['Ratings'] / 5) * 100
    
    # 5. Availability Hygiene (100% if Available, 0% if Not Available)
    result_df['Availability_Hygiene'] = (result_df['Availability'] == 'Yes') * 100
    
    # 6. Deal Hygiene (100% if Coupon Validation is True, 0% if False)
    result_df['Deal_Hygiene'] = result_df['Coupon Validation'] * 100
    
    # 7. Overall Brand Score
    result_df['Overall_Brand_Score'] = (
        (result_df['Price_Hygiene'].fillna(0) * 0.1) +
        (result_df['Activation_Hygiene'].fillna(0) * 0.05) +
        (result_df['Deal_Hygiene'].fillna(0) * 0.05) +
        (result_df['Availability_Hygiene'].fillna(0) * 0.2) +
        (result_df['EDD_Hygiene'].fillna(0) * 0.1) +
        (result_df['Rating_Hygiene'].fillna(0) * 0.2) +
        (result_df['Catalog_Hygiene'].fillna(0) * 0.2)+
        (result_df['Buy Box_Hygiene'].fillna(0) * 0.1)
    )   # Divide by 100 to get final percentage
     
    #Need to change the overall score calculation
    return result_df

df_input = calculate_hygiene_metrics(df_input)
```
